```python
import pandas as pd
from matplotlib import pyplot as plt
import datetime
```

```python
def get_close_plot(path):
    df = pd.read_csv(path)
    ts_volume = pd.Series(list(df['Volume']), index=df['Date'])
    return ts_volume
```

```python
ts_volume_intc = get_close_plot("./INTC.csv")
ts_volume_ibm = get_close_plot("./IBM.csv")
ts_volume_ms = get_close_plot("./MSFT.csv")
```

```python
ts_volume_nasdaq = get_close_plot("./^IXIC.csv")
ts_volume_nasdaq.index = ts_volume_nasdaq.index.to_datetime()
ts2 = ts_volume_nasdaq.groupby(pd.Grouper(freq='M'))
```

```python
ts2
```

```python
plt.plot(ts_volume_nasdaq.index, ts_volume_nasdaq.values)
plt.figure(figsize=(20,10))
plt.show()
```
